```python
m = teraz.minute
plt.figure(figsize=(12,12))
theta = np.linspace(0, np.pi*2) # okrąg
x0 = y0 = 0
r=1
x = x0 + r * np.cos(theta)
y = y0 + r * np.sin(theta)
plt.fill(x, y, facecolor="r", edgecolor='blue', linewidth=7)
plt.plot([0,0], [0.75, 1], color='k')
plt.plot([0.75,1], [0, 0], color='k')
plt.plot([0,0], [-0.75, -1], color='k')
plt.plot([-0.75,-1], [0, 0], color='k')
plt.plot(0,0, "ko")

h = teraz.hour
kat_h = -np.pi/2 + (h/12) * 2 * np.pi
x_h = x0 + r * np.cos(kat_h)
y_h = y0 + r * np.sin(kat_h)
plt.plot([0, x_h], [0, y_h], color="red")

plt.text(0, 0.75, '12', 
         horizontalalignment='center', 
         verticalalignment='top', 
         fontsize=18)
plt.text(0, -0.75, '6', 
         horizontalalignment='center', 
         verticalalignment='bottom', 
         fontsize=18)
plt.text(-0.75, 0, '9', 
         horizontalalignment='left', 
         verticalalignment='center', 
         fontsize=18)
plt.text(0.75, 0, '3', 
         horizontalalignment='right', 
         verticalalignment='center', 
         fontsize=18)
plt.axis('equal')
plt.show()
```

### Pasażerowie

```python
flights=sns.load_dataset("flights")
sredniPasazerowie = flights["passengers"].groupby(flights["year"]).agg(np.mean).reset_index()
sredniPasazerowie
```

```python
plt.figure(figsize=(8,6))
plt.plot(sredniPasazerowie["year"], sredniPasazerowie["passengers"], 'bo--')
plt.title("Wykres średniej liczby pasażerów od roku", color="m", fontsize=15)
plt.show()
```

```python
tips = sns.load_dataset("tips")
smoker = tips.smoker.value_counts()
smoker
```

```python
plt.pie(smoker, startangle=65, labels=['a', 'b'])

plt.title("Palacze")
plt.show()
```

```python
smoker = tips[["smoker"]].groupby("smoker", as_index=False).size().reset_index()
plt.bar(smoker["smoker"], height=smoker[0])
plt.show()
```
