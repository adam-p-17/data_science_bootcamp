```python
movies = pd.read_csv(file, 
                     delimiter='::', 
                     header=None, 
                     names=['MovieID', 'Title', 'Genres'], 
                     engine='python')
users = pd.read_csv('ml-1m/users.dat', 
                     delimiter='::', 
                     header=None, 
                     names=['UserID', 'Gender', 'Age', 'Occupation', 'Zip-code'], 
                     engine='python')
ratings = pd.read_csv('ml-1m/ratings.dat', 
                     delimiter='::', 
                     header=None, 
                     names=['UserID', 'MovieID', 'Rating', 'Timestamp'], 
                     engine='python')
```

### jaki jest najlepszy film wszechczasów, (najlepszy, czyli ma najwyższą średnią ocenę) - to zadanie możesz rozwiązać wykonując złączenie(join) zbioru movies i ratings

```python
pivot = pd.pivot_table(ratings, index=['MovieID'], values=['Rating'], aggfunc="mean")
max_pivot = pivot[pivot['Rating'] == pivot.max(axis=0)[0]]
pd.merge(left=max_pivot, right=movies, left_on='MovieID', right_on='MovieID')
```

```python
data = pd.merge(left=movies, right=ratings)
```

```python
pd.pivot_table(data, values="Rating", index=["Title", "Genres"], aggfunc='mean').sort_values(by="Rating",
                                                                                            ascending=False)
```

### wykonaj poprzedni punkt, odrzucając wcześniej filmy które nie uzyskały wystarczająco dużo głosów

```python
data = ratings.groupby(["MovieID", "Rating"]).size().reset_index()
```

```python
data = data[data[0] > 100]
```

```python
pivot = pd.pivot_table(ratings[ratings["MovieID"].isin(data["MovieID"])], index=['MovieID'], values=['Rating'], aggfunc="mean")
```

```python
max_pivot = pivot[pivot['Rating'] == pivot.max(axis=0)[0]]
pd.merge(left=max_pivot, right=movies, left_on='MovieID', right_on='MovieID')
```

```python
movie_id = data.groupby(["MovieID", "Rating"]).size().reset_index()
movie_id = movie_id[movie_id[0] > 100][["MovieID"]]
pd.pivot_table(data[data["MovieID"].isin(movie_id["MovieID"])], values="Rating", index=["Title", "Genres"], aggfunc='mean')\
        .sort_values(by="Rating", ascending=False)[:1]
```

### jaki jest najlepszy film według kobiet i według mężczyzn

```python
ratings_user = pd.merge(left=ratings, right=users)
```

```python
srednieOcenyPlcie = ratings_user.groupby(["MovieID", "Gender"]).agg({"Rating": ["mean", "size"]})
srednieOcenyPlcie.columns = ["mean", "size"]
srednieOcenyPlcie = srednieOcenyPlcie.reset_index()
srednieOcenyPlcie100 = srednieOcenyPlcie[srednieOcenyPlcie["size"] > 100]
pd.merge(srednieOcenyPlcie100, movies).sort_values(by="mean", ascending=False)\
                                                    .drop_duplicates(subset="Gender")
```

### jaki jest średni rok oglądanego filmu w poszczególnych grupach wiekowych

```python
data = pd.merge(pd.merge(left=users, right=ratings), right=movies)
```

```python
pd.pivot_table(data, values="Year", index="Age", aggfunc="mean")
```

### jakie trzy gatunki filmowe są najczęściej oglądane przez kobiety i mężczyzn

```python
result = movies["Genres"].str.split("|", expand=True)
data = pd.concat([movies, result], axis=1)
```

```python
data.head()
```

```python
result = pd.melt(data, id_vars=["MovieID", "Title", "Year"], value_vars=[0,1,2,3,4,5])
```

```python
df_2 = pd.merge(left=users, right=ratings)
df_2 = pd.merge(left=df_2, right=result)
```

```python
pd.pivot_table(df_2[df_2["Gender"] == 'F'], values="Rating", index=["value"], aggfunc="count")\
                .sort_values(by="Rating", ascending=False)[:3]
```

```python
pd.pivot_table(df_2[df_2["Gender"] == 'M'], values="Rating", index=["value"], aggfunc="count")\
                .sort_values(by="Rating", ascending=False)[:3]
```
