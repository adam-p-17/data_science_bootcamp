## tabela kontyngencji

```python
result = pd.DataFrame(planes.engine.value_counts())
result = result.reset_index().sort_values(by="index")[["engine", "index"]].reset_index(drop=True)
# result.columns = ["COUNT(*)", "engine"]
result = result.rename({"engine": "COUNT(*)", "index": "engine"}, axis=1)
result
```

## 4

```python
query = "SELECT COUNT(*), engine, type FROM planes GROUP BY engine, type"
sql_result = pd.read_sql_query(query, con)
sql_result
```

## Tabela przestawna

```python
result = pd.pivot_table(planes[["engine", "type"]], index=['engine', 'type'], aggfunc='size').reset_index()\
    [[0, "engine", "type"]]
result.columns = ["COUNT(*)", "engine", "type"]
result
```

## Group by

```python
result2 = pd.DataFrame(planes[["engine", "type"]].groupby(["engine", "type"], as_index=False).size())
result2.reset_index(drop=False).rename({0: "COUNT(*)"}, axis=1)[["COUNT(*)", "engine", "type"]]
```

```python
result3 = pd.crosstab(planes.engine, planes.type).reset_index()
result3 = result3.melt(id_vars=["engine"])
result3 =result3[result3.value > 0][["value", "engine", "type"]]
result3.columns = ["COUNT(*)", 'engine', 'type']
result3.sort_values(by="engine")
```

## 14

```python
query = 'SELECT manufacturer, COUNT(*) AS howmany FROM planes GROUP BY manufacturer ORDER BY howmany'
sql_result = pd.read_sql_query(query, con)
sql_result
```

```python
planes.groupby("manufacturer").size().reset_index().rename({0: "howmany"}, axis=1).sort_values(by="howmany").reset_index(drop=True)
```

## 15

```python
query = 'SELECT manufacturer, COUNT(*) AS howmany FROM planes GROUP BY manufacturer ORDER BY howmany DESC LIMIT 10'
sql_result = pd.read_sql_query(query, con)
sql_result
```

```python
result = planes.groupby("manufacturer").size().reset_index().rename({0: "howmany"}, axis=1)
result.sort_values(by="howmany", ascending=False).reset_index(drop=True).head(10)
```

```python
pd.concat([A, B])
```

## SELECT * FROM A UNION SELECT * FROM B

```python
result = pd.concat([A, B]).drop_duplicates()
result
```

## SELECT * FROM A INTERSECT SELECT * FROM B

```python
result = pd.concat([A, B])
result[result.duplicated()]
```

## SELECT * FROM A EXCEPT SELECT * FROM B

```python
A[~A.isin(B)].dropna()
```

## SELECT * FROM B EXCEPT SELECT * FROM A
