```python
import pandas as pd
import requests
import re
```

```python
url = 'https://pl.wikipedia.org/wiki/Zygmunt_III_Waza'
r = requests.get(url)
ahref = re.findall("<a .*?href=[\"'](.+?)[\"'].*?>(.*?)</a>", r.text)
```

```python
pd.DataFrame(ahref).rename({0: "referer", 1: "text"}, axis=1)
```

```python
url = 'https://stooq.pl/q/?s=cdr'
r = requests.get(url)
```

```python
kurs = float(re.search('<.*?aq_cdr_c2.*?>(.*?)</span>', r.text).groups()[0])
data = re.search('<.*?aq_cdr_d2.*?>(.*?)</span>', r.text).groups()[0]
zm_bezwzgledna = float(re.search('<.*?aq_cdr_m2.*?>(.*?)</span>', r.text).groups()[0])
zm_wzgledna = float(re.search('<.*?aq_cdr_m3.*?>\((.*?)%\)</span>', r.text).groups()[0])
```

```python
ramka = pd.DataFrame([{"Kurs": kurs, 
                      "Data": data, 
                      "Zmiana Bezwzględna": zm_bezwzgledna, 
                      "Zmiana Względna": zm_wzgledna}])
```
