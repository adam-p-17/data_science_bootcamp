```python
import re
pattern = r"Z(\d+)\((\d+)\+(\d+)\)"
```

```python
df2 = df.copy()
```

```python
df2 = df2.rename({"Z13(5)": "Z13(4+1)"}, axis=1)
for col in df2.columns:
    if '+' in col:
        df2 = pd.concat([df2, df2[col].str.split("+", expand=True)], axis=1)
        regex = re.findall(pattern, col)
        df2 = df2.rename({0: f'Z{regex[0][0]}({regex[0][1]})', 1: f'L{regex[0][0]}({regex[0][2]})'}, axis=1)
```

```python
df2 = df2.drop(columns = list(df2.columns[2:17]))
```

```python
df2 = df2.replace(["NA", r"NS.+", "NB"], 0, regex=True)
df2 = df2.replace(["US", "ZW"], np.NaN)
```

```python
df2 = pd.concat([df2.iloc[:, 0:3], df2.iloc[:, 3:].astype(float)], axis=1)
df2.dtypes
df2.fillna(df2.mean(numeric_only=True, skipna=True), axis=0)
```

```python
for columnIndex in range(3, 30):
    for rowIndex in range(df2.shape[0]):
        if np.isnan(df2.iloc[rowIndex, columnIndex]):
            df2.iloc[rowIndex, columnIndex] = np.mean(np.r_[df2.iloc[:, columnIndex].mean(),\
                              df2.iloc[rowIndex, 3:31].mean()]) 
df2
```

```python
df2.sum(axis=1)
```

```python
sumaPunktow = df2.columns.str.extract(r'.*\((\d+)\).*').astype(float).iloc[:,0].sum()
```

```python
df2[f'suma({sumaPunktow})'] = df2.sum(axis=1)
```

```python
df2["procent"] = df2["suma(69.0)"] / sumaPunktow
```

```python
df["ocena"] = pd.cut(df["procent"], [0, 0.5, 0.6, 0.7, 0.8, 0.9], labels=[2,3,3.5,4.5,5])
```
