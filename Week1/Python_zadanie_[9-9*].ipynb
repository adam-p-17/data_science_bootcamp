```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def srednie_zarobki(zarobki):
    srednia = int(round(np.mean(zarobki)))
    return f'Średnie zarobki działu wynoszą {srednia}'
```

```python
zarobki = [random.gauss(3200, 1000) for i in range(1500)]
```

```python
srednie_zarobki(zarobki)
```

```python
plt.hist(x = zarobki)
plt.xlabel('Zarobki w PLN')
plt.ylabel('Liczebność w .tys')
plt.title('Rozkład zarobków w dziale X', color='blue', fontsize=20)
plt.show()
```
